```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
```

```python
# note that to use tfds builder, the file must first be downloaded manually from cityscapes-dataset.com first
# then place the .zip files into ./tensorflow_datasets/downloads/manual
cityscape_builder = tfds.builder('cityscapes')
cityscape_builder.download_and_prepare()
```

```python
cityscapes = cityscape_builder.as_dataset()
```

```python
cityscapes['train']
```

```python
sample = tfds.as_dataframe(cityscapes['train'].take(10),)
sample.head()
```

```python
# showing one example
x = sample.loc[0]
img,mask  = x[1],x[2]

fig = plt.figure(figsize=(15,8))
fig.add_subplot(1,2,1)
plt.imshow(img)
plt.title('streetview image')

fig.add_subplot(1,2,2)
plt.imshow(mask)
plt.title('mask')

```
