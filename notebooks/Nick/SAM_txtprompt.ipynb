```python
import torch
import torchvision
import PIL

class SAM_segmenter:

    def __init__(self):
      !pip install -U git+https://github.com/luca-medeiros/lang-segment-anything.git
      from lang_sam import LangSAM
      self.model = LangSAM()

    def segment(self, image, prompt):
      image = PIL.Image.fromarray(np.uint8(image), mode='RGB')
      mask = self.model.predict(image, prompt)
      return mask

    def segment_all(self, image):
      prompts = [
          'sidewalks',
          'building',
          'wall',
          'fence',
          'guard rail',
          'bridge',
          'tunnel',
          'vegetation',
          'poles'
      ]
      output = list()
      for i in prompts:
        mask = self.segment(image, prompt=i)
        output.append(mask)
      return output
```

```python
Model = SAM_segmenter()
```
